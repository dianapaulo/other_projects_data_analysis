# barchart with matplotlib calling .plot(kind="bar") 

```python
#we can plot a barchart with matplotlib calling .plot(kind="bar") directly on the Dataframe.
#plot the top 10 neighbourhood with Airbnb listings
ab_NYC_df["neighbourhood"].value_counts().head(10).plot(kind="bar")
```

# Using Seaborn 

```python
#no need for a value_counts() with seaborn we can just use a countplot
#seaborn for data visualization based on matplotlib
import seaborn as sns

sns.countplot(data = ab_NYC_df, x = "neighbourhood_group")
```

```python
#put in order for a better look
order = ab_NYC_df["neighbourhood_group"].value_counts().index
sns.countplot(data = ab_NYC_df, x = "neighbourhood_group", order = order)
```

```python
#What is the influence neighbourhood group on room type?
#FYI, here are the room types:
ab_NYC_df["room_type"].unique()
```

```python
sns.countplot(data = ab_NYC_df, x = "neighbourhood_group", hue = "room_type")
```

# Using  displot:

```python
#What is the distribution of listing prices?
#We can use a displot:
sns.displot(ab_NYC_df["price"])
```

```python
#showing the affordable listings <= 500
affordable_ab_NYC = ab_NYC_df[ab_NYC_df["price"] <= 500]
```

```python
ab_NYC_df.shape #<=500
```

```python
#much more imformative chart
sns.displot(affordable_ab_NYC["price"])
```

```python
#more broad using bins, the bins parameter enables you to control the bins of the histogram (i.e., the number of bars).
sns.displot(affordable_ab_NYC["price"], bins = 10)
```

```python
#We can use Pandas for the mean:
ab_NYC_df.price.mean()
```

```python
affordable_ab_NYC.price.mean()
```

# Violinplot

```python
#What is the distribution of the listing prices based on the Neighboorhood <group?
#using  violinplot
sns.violinplot(data=affordable_ab_NYC, x ="neighbourhood_group", y="price")
```

```python
#we can make the figure bigger calling before the sns.violinplot(...)
#The white dot is the mean value
from matplotlib import pyplot as plt
plt.figure(figsize=(15,8))
sns.violinplot(data=affordable_ab_NYC, x ="neighbourhood_group", y="price")
```

# Matplotlib scatterplot

```python
#to find out how far east or west a place is, lines of longitude are used

from matplotlib import pyplot as plt
affordable_ab_NYC.plot(
    kind = "scatter",
    x = "longitude",
    y = "latitude",
    c = "price",
    cmap = "inferno",
    colorbar = True,
    alpha = 0.8,
    figsize = (12,8))
```

# MAP with image

```python
#we can see the density of Airbnb in New York City. Orange and yellow are the expensive ones
background_image = plt.imread("NY_map.png")
plt.imshow(background_image, zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])

ax = plt.gca()
affordable_ab_NYC.plot(
    ax = ax,
    zorder = 1,
    kind = "scatter",
    x = "longitude",
    y = "latitude",
    c = "price",
    cmap = "inferno",
    colorbar = True,
    alpha = 0.8,
    figsize = (12,8))
```
